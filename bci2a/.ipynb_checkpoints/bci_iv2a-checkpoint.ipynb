# Chapter One: About the dataset
This dataset contains motor imagery EEG-BCI recordings from 9 subjects in multiple sessions (one file for training, and another file for testing)
In the experiments, the subject was asked to imagine perfoming the following four types of movement: left hand, right hand, tongue, and foot.
Each imagination of a movement is called a trial and the task is to classify all givens trials to their corresponding movement types.
We will first try to visualize some trials to better understand the type of data 



```python
from pathlib import Path
import scipy.io
# Load subject 1's training data

data = scipy.io.loadmat("/vsc-hard-mounts/leuven-data/379/vsc37996/test/A01T.mat")
data = data['data']
num_sessions = len(data[0])
print(f"There are {num_sessions} sessions")

this_rec_session_data = {}
for thisSession in range(num_sessions):
    this_rec_session_data[thisSession] = {name: data[0][thisSession][name][0, 0] for name in data[0][thisSession].dtype.names}
```

```python
# Now we have a nice structure that contains a lot of info, we can check 
this_rec_session_data[3]
# X is the data,trial is the index of that trial, y is the label, fs is the sampling rate, classes are the class descriptions, 
# artifacts is 0/1 indicating if there is any artifact in this trial, gender and age are meta data for this subject   
```

```python
# Now we can visualize the data using mne, mne is a python package useful for EEG analysis that contains a lot of handy functions, 
# but first, we need to create a raw object
import mne
import numpy as np
mne.set_log_level('ERROR')

this_session_data = this_rec_session_data[3]

# 1) Inspect the raw array shape first
print("X shape:", np.array(this_session_data["X"]).shape)

# 2) Build channels  time explicitly (22 EEG channels)
#    Common BCI IV-2a loaders give X as (n_times, n_channels) or (n_trials, n_times, n_channels).
X = np.array(this_session_data["X"])

# Case A: (n_times, n_channels)
if X.ndim == 2 and X.shape[1] >= 22:
    eeg_uV = X[:, :22].T  # -> (22, n_times)

# Case B: (n_trials, n_times, n_channels)  pick one trial or stitch them
elif X.ndim == 3 and X.shape[2] >= 22:
    # Example: visualize the first trial only (or np.concatenate across trials)
    eeg_uV = X[0, :, :22].T  # -> (22, n_times)

else:
    raise ValueError("Unexpected X shape; need at least 22 channels on the last axis.")

print("eeg_uV shape (should be 22 x n_times):", eeg_uV.shape)

# 3) Convert V  V (if your values look like tens/hundreds: theyre probably V)
eeg_V = eeg_uV.astype(float) * 1e-6

# 4) Build Raw
fs = float(this_session_data["fs"][0][0])
ch_names = ['Fz','FC3','FC1','FCz','FC2','FC4','C5','C3','C1','Cz','C2','C4',
            'C6','CP3','CP1','CPz','CP2','CP4','P1','Pz','P2','POz']
info = mne.create_info(ch_names=ch_names, sfreq=fs, ch_types='eeg')
raw = mne.io.RawArray(eeg_V, info, verbose=False)# eeg_V  info 
raw.set_montage("standard_1005", match_case=False)#montage

# Optional: quick clean to improve readability
raw.set_eeg_reference('average', projection=False)
raw.notch_filter(50.0, verbose=False)#5050
raw.filter(1., 40., fir_design='firwin', verbose=False)#1Hz40Hz

# 5) Make sure the browser actually stacks 22 channels
#    (In some Jupyter backends it can look odd if it falls back to a static figure.)
try:
    mne.viz.set_browser_backend('qt')  # if Qt is available this gives the best browser
except Exception:
    pass

raw.plot(
    n_channels=22,                 # show all
    duration=5.0,                  # seconds per page
    scalings=dict(eeg=10e-6),      # 50 V per channel
)
```

What we see is an entire recording lasts for 5 minutes, since we are only interested in the task trials, we need to segment this recording and find the corresponding trials. # Chapter Two: Data pre-processing
Typically, EEG needs to undergo certain pre-processing steps before decoding. Below we use a simple function to demonstrate it.
This function takes in eeg (mne raw object), and performs notch filtering to remove the power line noise, and a bandpass filter between 0.5-64 Hz, and downsampled to 128 Hz

```python
def process_eeg(eeg, f_l, f_h, fs, info):
    # common average reference
    #eeg.set_eeg_reference('average', projection=True)
    freqs = [50]  # Frequencies to notch filter
    eeg = eeg.notch_filter(freqs=freqs)#
    # band pass between 0.5-64 to remove slow drifts and high freq noise
    eeg.filter(f_l, f_h, picks=info['ch_names'], fir_design='firwin')
    # down sample 
    eeg.resample(fs, npad="auto")
    return eeg._data

data = process_eeg(raw, 0.5, 64, 128, info)#128
print(data.shape)
```

# Chapter Three: Data fetching for ML
A good news is that I have performed all steps for you. You can even just use the following function and classes to get a dataset class that is ready to be used for machine learning.

```python
import torch
import numpy as np
import pickle
import os
import time
import pandas as pd

class BCI2aDataset(torch.utils.data.Dataset):
    def __init__(
        self,
        data_path: str,#.pkl
        fold: int = 0,#
        classification_task: str = "bci_iv2a", 
        data_length: int = 512,     # window length in samples1000512
        train: bool = True,         #
        class_label: dict = None,            # maps target  int
        transform=None,                      # any data transformation  augmentation
        chan_info: torch.Tensor = None       # channel info, needed for foundation models
        # (Foundation Models) EEG
    ):
        # basic attrs
        self.subjectName      = os.path.splitext(os.path.basename(data_path))[0]
        self.fold             = fold
        self.task             = classification_task
        self.data_length      = data_length
        self.train            = train
        self.class_label      = class_label or {}
        self.transform        = transform
        self.chan_info        = chan_info

        # load pickled data
        with open(data_path, "rb") as f:
            sub = pickle.load(f)
            
        if self.train:
            self.data = sub["trainX"]
            self.labels = sub["trainY"]
        else:
            self.data = sub["testX"]
            self.labels = sub["testY"]    
            
        if self.transform:
            self.data = self.transform(self.data)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        trial_data = self.data[idx,:,:]
        label = self.labels[idx]
        
        if self.data_length:
            trial_data = trial_data[:,:self.data_length]
            
        if self.chan_info is not None:
            return  torch.from_numpy(trial_data).type(torch.FloatTensor),  self.class_label[label], self.chan_info
        else:
            return  torch.from_numpy(trial_data).type(torch.FloatTensor),  self.class_label[label]
```

```python
class_label = {"left_hand": 0, "right_hand": 1, "feet": 2, "tongue": 3}
train_dataset = BCI2aDataset(data_path="/lustre1/project/stg_00160/benchmark_preprocessed_datasets/bci_iv2a/A1.pkl",fold=1,data_length=int(4*128),train=True,class_label=class_label)
test_dataset = BCI2aDataset(data_path="/lustre1/project/stg_00160/benchmark_preprocessed_datasets/bci_iv2a/A1.pkl",fold=1,data_length=int(4*128),train=False,class_label=class_label)
```

# Chapter Four: Training an EEGNet
EEGNet is a classic CNN based EEG BCI decoder. We will use it for demonstration.
Also notice, we are building a subject-specific model, a model that is both trained and tested on the same subject

```python
"""
EEGNet: 

EEGNet: A Compact Convolutional Network for EEG-based Brain-Computer Interfaces

"""

# down sample 128Hz
import torch.nn as nn
import torch
import torch.nn.functional as F
import math

class SeparableConv2d(nn.Module):

    def __init__(self, in_channels, out_channels, kernel_size, bias=False):
        super(SeparableConv2d, self).__init__()
        self.depthwise = nn.Conv2d(in_channels, in_channels, kernel_size=kernel_size, 
                                   groups=in_channels, bias=bias,padding='same')
        self.pointwise = nn.Conv2d(in_channels, out_channels, 
                                   kernel_size=1, bias=bias)

    def forward(self, x):
        out = self.depthwise(x)
       # print("sepcov2d-depth",out.shape)
        out = self.pointwise(out)
       # print("sepcov2d-point",out.shape)
        return out
    
class MaxNormLinear(nn.Module):
    
    def __init__(self,inchannel,outchannel):
        
        super(MaxNormLinear, self).__init__()
        self.linear = nn.Linear(inchannel, outchannel)
        self._eps = 1e-7
        
    def max_norm(self):
        with torch.no_grad():
            norm = self.linear.weight.norm(2, dim=0, keepdim=True)
            desired = torch.clamp(norm, 0, 0.25)
            self.linear.weight = torch.nn.Parameter( (self.linear.weight * (desired / (self._eps + norm))), dtype=torch.float, device="cuda")
    
    def forward(self,x):
        self.max_norm()
        return self.linear(x)
        
class ConstrainedConv2d(nn.Module):
    def __init__(self,in_channels, out_channels, kernel_size, no_groups, if_bias):      
        super(ConstrainedConv2d, self).__init__()
        self.conv2 = nn.Conv2d(in_channels, out_channels, kernel_size,
                               groups=no_groups, bias=if_bias)  # spatial filter 22channel=>3channel
    def max_norm(self):
       
        norm = self.conv2.weight.norm(2, dim=0, keepdim=True)
        desired = torch.clamp(norm, 0, 1.0)
        self.conv2.weight = torch.nn.Parameter( (self.conv2.weight * (desired / (1e10-7 + norm)))).to("cuda")
    
    def forward(self, input):
        self.max_norm()
        return self.conv2(input)  
    
class EEGNet(nn.Module):
    def __init__(self, no_spatial_filters, no_channels, no_temporal_filters, temporal_length_1, temporal_length_2, window_length, num_class, drop_out_ratio=0.50, pooling2=4, pooling3=8):
        super(EEGNet, self).__init__()
        self.drop_out_ratio = drop_out_ratio
        
        # Layer 1: band pass filter
        self.conv1 = nn.Conv2d(1, no_temporal_filters, (1, temporal_length_1), padding='same', bias=False)

        self.batchnorm1 = nn.BatchNorm2d(no_temporal_filters, False)
        self.dropout = nn.Dropout(self.drop_out_ratio)
        
        # Layer 2: channel-aware spatial filter
        self.conv2 = nn.Conv2d(no_temporal_filters, no_temporal_filters * no_spatial_filters, (no_channels, 1),
                               groups = no_temporal_filters, bias = False)  # spatial filter 
        self.batchnorm2 = nn.BatchNorm2d(no_temporal_filters * no_spatial_filters, False)
        self.pooling2 = nn.AvgPool2d(1, pooling2) # from fs->32 Hz
        
        # Layer 3
        self.separableConv2 = SeparableConv2d(no_temporal_filters * no_spatial_filters,
                                              no_temporal_filters * no_spatial_filters, (1, temporal_length_2))
        self.batchnorm3 = nn.BatchNorm2d(no_temporal_filters * no_spatial_filters, False)

        self.pooling3 = nn.AvgPool2d((1, pooling3)) 
        
        eeg_random = torch.randn(4,no_channels,window_length)
        fc_length = self.calc_fc_features(eeg_random)
        self.fc1 = nn.Linear(fc_length, num_class)
        
    
    def calc_fc_features(self,x):#fc1
        self.eval()
        with torch.no_grad():
            x = torch.unsqueeze(x,1)
            x = self.conv1(x)
            x = self.batchnorm1(x)
            x = self.conv2(x)
            B,FB,Ch,TL = x.shape
            x= torch.reshape(x,(B,FB*Ch,1,TL))
            x = nn.functional.elu(self.batchnorm2(x))
            x = self.pooling2(x)
            x = self.dropout(x)
            x = self.separableConv2.forward(x)
            x = nn.functional.elu(self.batchnorm3(x))
            x = self.pooling3(x)
            x = self.dropout(x)
            x = torch.flatten(x, start_dim=1)
            return x.shape[-1]
    
    def set_drop_out(self,new_dropout):#
        self.dropout.rate = new_dropout
    
            
    def forward(self, x):
        # Layer 1
        #print("input",x.shape)
        x = torch.unsqueeze(x,1)#Conv2d(, 1, , )
        #print(x.shape)
        x = self.conv1(x)# (1, 64)8
        #print(x.shape)
        x = self.batchnorm1(x)
        # Layer 2
        #print("cov1",x.shape)
        x = self.conv2(x)# (22, 1) EEG 
        #groups = no_temporal_filters:  PyTorch 1
        #print("conv2", x.shape)
        B,FB,Ch,TL = x.shape
        x= torch.reshape(x,(B,FB*Ch,1,TL))
        x = nn.functional.elu(self.batchnorm2(x))
        x = self.pooling2(x)
        x = self.dropout(x)

        # Layer 3
        #print("pooling2",x.shape)
        x = self.separableConv2.forward(x)#
        #print("cov3",x.shape)

        x = nn.functional.elu(self.batchnorm3(x))
        x = self.pooling3(x)
        #print("pooling3",x.shape)
        x = self.dropout(x)
        #print("before flatten",x.shape)
        x = torch.flatten(x, start_dim=1)#
        #print("fc",x.shape)
        x = self.fc1(x)#
        return x
```

```python
eegnet = EEGNet(no_spatial_filters=4, no_channels=22, no_temporal_filters=8, temporal_length_1=64, temporal_length_2=16, window_length=int(4*128), num_class=4, drop_out_ratio=0.50, pooling2=4, pooling3=8)
```

```python
import torch
from torch.utils.data import DataLoader

# --- config ---
batch_size   = 8
num_workers  = 0
epochs       = 50
lr           = 1e-3
weight_decay = 1e-4

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = eegnet.to(device)

# --- data loaders ---
train_loader = DataLoader(
    train_dataset, batch_size=batch_size, shuffle=True,
    num_workers=num_workers, pin_memory=True, drop_last=False
)
test_loader = DataLoader(
    test_dataset, batch_size=batch_size, shuffle=False,
    num_workers=num_workers, pin_memory=True, drop_last=False
)

# --- loss & optimizer ---
criterion = torch.nn.CrossEntropyLoss()          # expects logits and int64 labels
optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=weight_decay)

# --- helpers ---
def run_epoch(loader, train: bool):
    if train:
        model.train()
    else:
        model.eval()

    total_loss, total_correct, total_samples = 0.0, 0, 0

    for x, y in loader:
        x = x.to(device, non_blocking=True)
        y = y.to(device, non_blocking=True).long()   # ensure class labels are int64

        if train:
            optimizer.zero_grad(set_to_none=True)

        # forward
        logits = model(x)                            # shape [B, C]
        loss = criterion(logits, y)

        if train:
            loss.backward()
            optimizer.step()

        # stats
        total_loss += loss.item() * x.size(0)
        preds = logits.argmax(dim=1)
        total_correct += (preds == y).sum().item()
        total_samples += x.size(0)

    avg_loss = total_loss / max(total_samples, 1)
    acc = total_correct / max(total_samples, 1)
    return avg_loss, acc

# --- training loop ---
best_test_acc = 0.0
for epoch in range(1, epochs + 1):
    train_loss, train_acc = run_epoch(train_loader, train=True)
    test_loss, test_acc   = run_epoch(test_loader,  train=False)

    if test_acc > best_test_acc:
        best_test_acc = test_acc
        torch.save({'model': model.state_dict()}, 'best_model.pt')

    print(
        f"Epoch {epoch:03d} | "
        f"train_loss: {train_loss:.4f}  train_acc: {train_acc*100:.2f}% | "
        f"test_loss: {test_loss:.4f}    test_acc: {test_acc*100:.2f}%"
    )

print(f"Best test acc: {best_test_acc*100:.2f}%")
```
